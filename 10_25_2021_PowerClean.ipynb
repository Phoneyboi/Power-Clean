# CONVERSION OF DATA TO CREATE A SINGLE TOTAL MAGNITUDE
Below is the code to convert the acceleration of the 'x' and 'z' axes to a single total magnitude. This is crucial in the analysis of the movement of the bar because it converts the acceleration of two single axes, into one total acceleration of the bar. 

It is hard to explain through text what is actaully going on with the data and the physical movement of the bar, but essentially, the linear movement of the bar from the floor to the shoulders can be characterized by the acceleration of the 'x' and 'z' axes. Combining the acceleration of those two axes into one magnetude shows the total linear acceleration of the bar during the lift.

```python
# This cell takes in the following columns of data: 'az(g)', 'ax(g)', 'AngleY'.
# These three columns of data will be used to calculate the total acceleration of the bar at descrete points.

xmin = 0
xmax = len(dframe1)
atotal = []

while xmin < xmax:
    az = dframe1['az(g)'][xmin]*np.cos(np.deg2rad(dframe1['AngleY(deg)'][xmin]))
    ax = dframe1['ax(g)'][xmin]*np.sin(np.deg2rad(dframe1['AngleY(deg)'][xmin]))
    # Converting data, given as degrees, into radians
    atot = np.absolute(az) + np.absolute(ax)
    atotal.append(atot)
    xmin = xmin + 1
    
#print(atotal)
```

```python
dframe1['atotal'] = atotal
```

The plot below shows the total acceleration of the bar as a function of time.  

```python
sns.scatterplot(data=dframe1[:], x='ChipTime', y='atotal')
```
